```python
import pandas as pd
import re
from time import strptime
import time
from datetime import datetime
```

```python
allData = pd.read_csv("LAHacks Deidentified.csv")
```

```python
allData.head()
```

```python
allData.shape
```

```python
allData= allData[allData['Category Name'] != 'Unassigned']
allData['Month'] = [re.sub(r'[0-9]|-','',x) for x in allData['Shipping Date'].tolist()]
allData['Month'] = [strptime(x, '%b').tm_mon for x in allData['Month'].tolist()]
allData['Day'] = [int(x[0:2]) for x in allData['Shipping Date'].tolist()]
allData['Year'] = [int('20'+str(x[len(x)-2:len(x)])) for x in allData['Shipping Date'].tolist()]

allData.head()
```

```python
datetime_arr = [] 
for x in range(len(allData)):
#     print(allData['Year'].iloc[x])
#     print(allData['Month'].iloc[x])
#     print(allData['Day'].iloc[x])
    datetime_arr.append(datetime(allData['Year'].iloc[x], allData['Month'].iloc[x], allData['Day'].iloc[x]))
#allData['datetime'] = allData[['Year','Month', 'Day']].apply(datetime)
```

```python
allData['datetime'] = datetime_arr
allData['datetime'].iloc[0]
allData = allData.drop(allData.columns[[11, 12, 13]], axis=1)
allData['numeric_datetime'] = [time.mktime(i.timetuple()) for i in allData['datetime']]
allData.head()
```
